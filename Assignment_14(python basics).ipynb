```python
#3
The box tuple is used to define the rectangular area within an image ,it is a tuple of 4 integers
representing the coordinates:(left,upper,right,lower)

How it works:

Left: The x-coordinate of the left edge.

Upper: The y-coordinate of the top edge.

Right: The x-coordinate of the right edge.

Lower: The y-coordinate of the bottom edge.
```

```python
#8
ImageDraw objects contain the drawing methods in Pillow. I can create an ImageDraw object using
the ImageDraw.Draw() function, passing it an Image object.
```

```python
from PIL import Image, ImageDraw

# Create a new image or open an existing one
image = Image.new('RGB', (200, 200), 'white')

# Get an ImageDraw object
draw = ImageDraw.Draw(image)

# Now you can use draw methods, like drawing a rectangle
draw.rectangle((50, 50, 150, 150), outline='black', fill='blue')

# Save or display the image
image.save('path/to/your/shape_drawing_image.jpg')
image.show()
```
